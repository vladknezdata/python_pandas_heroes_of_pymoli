```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# District Summary Values

districtSummary = pd.DataFrame()
districtSummary['Total Schools'] = [school_data_complete['School ID'].nunique()]
districtSummary['Total Students'] = school_data_complete['Student ID'].count()
districtSummary['Total Budget'] = school_data['budget'].sum()
districtSummary['Average Math Score'] = school_data_complete['math_score'].mean()
districtSummary['Average Reading Score'] = school_data_complete['reading_score'].mean()
districtSummary['% Passing Math'] = \
    school_data_complete[school_data_complete['math_score'] >= 70].shape[0] / districtSummary['Total Students'] * 100
districtSummary['% Passing Reading'] = \
    school_data_complete[school_data_complete['reading_score'] >= 70].shape[0] / districtSummary['Total Students'] * 100
districtSummary['% Overall Passing Score'] = (districtSummary['% Passing Math'] + districtSummary['% Passing Reading']) / 2

# cell formating
districtSummary['Total Students'] = districtSummary['Total Students'].map("{:,}".format)
districtSummary['Total Budget'] = districtSummary['Total Budget'].map("${:,.2f}".format)
districtSummary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
topSchools = pd.DataFrame()

topSchools['Total Students'] = school_data.groupby(['school_name', 'type'])['size'].sum()
topSchools['Total School Budget'] = school_data.groupby(['school_name', 'type'])['budget'].mean()
topSchools['Per Student Budget'] = topSchools['Total School Budget'] / topSchools['Total Students']
topSchools['Average Math Score'] = school_data_complete.groupby(['school_name', 'type'])['math_score'].mean()
topSchools['Average Reading Score'] = school_data_complete.groupby(['school_name', 'type'])['reading_score'].mean()
topSchools['% Passing Math'] = school_data_complete[school_data_complete['math_score'] >= 70].groupby(['school_name', 'type'])['math_score'].count() / topSchools['Total Students'] * 100
topSchools['% Passing Reading'] = school_data_complete[school_data_complete['reading_score'] >= 70].groupby(['school_name', 'type'])['reading_score'].count() / topSchools['Total Students'] * 100
topSchools['% Overall Passing Rate'] = (topSchools["% Passing Math"] + topSchools["% Passing Reading"]) /2
topSchools = topSchools.sort_values("% Overall Passing Rate", ascending=False)
topSchools.head()
```

```python
l = list(school_data_complete['grade'].unique())
```

```python
byGradeMath = pd.DataFrame()
for i in range(len(l)):
    byGradeMath[l[i]] = school_data_complete[school_data_complete['grade'] == l[i]].groupby(['school_name'])['math_score'].mean()
byGradeMath = byGradeMath[['9th', '10th', '11th', '12th']]
byGradeMath
```

```python
byGradeReading = pd.DataFrame()
for i in range(len(l)):
    byGradeReading[l[i]] = school_data_complete[school_data_complete['grade'] == l[i]].groupby(['school_name'])['reading_score'].mean()
byGradeReading = byGradeReading[['9th', '10th', '11th', '12th']]
byGradeReading
```
